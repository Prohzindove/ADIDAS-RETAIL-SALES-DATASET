```sql
USE workspace.addidas_sales_dataset; --Import dataset

--View dataset
SELECT *
FROM adidas_us_sales_datasets
LIMIT 100;

--Restore wrongly deleted table
RESTORE TABLE adidas_us_sales_datasets
TO TIMESTAMP AS OF '2026-06-11 14:42:15';

--Rerun the dataset to see if the table has been restored
SELECT *
FROM adidas_us_sales_datasets
LIMIT 100;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c1 TO Retailer;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN `Adidas Sales Database` TO `Retailer ID`;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c3 TO Invoice_Date;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c4 TO Region;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c5 TO State;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c6 TO City;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c7 TO Product;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c8 TO Price_per_Unit;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c9 TO Units_Sold;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c10 TO Total_Sales;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c11 TO Operating_Profit;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c12 TO Operating_Margin;

ALTER TABLE adidas_us_sales_datasets
RENAME COLUMN _c13 TO Sales_Method;

--Clean the dataset from Null values
ALTER TABLE adidas_us_sales_datasets
DROP COLUMN _c0

DELETE FROM adidas_us_sales_datasets
WHERE _c1 IS NULL;

--Delete the first row that has been wrongly entered
DELETE FROM adidas_us_sales_datasets
WHERE Retailer = 'Retailer';

--Check to see if the data cleaning was effected
SELECT *
FROM adidas_us_sales_datasets
LIMIT 100;

--Show the profits starting with the highest
SELECT *
FROM adidas_us_sales_datasets
ORDER BY Operating_Profit DESC;

--Select maximum operating profit for each region
SELECT Region,
       MAX(Operating_Profit) AS MAX_Profit
FROM adidas_us_sales_datasets 
GROUP BY Region
ORDER BY MAX_Profit ASC; 

--Show sales in the Northeast region where the city is New York
SELECT * 
FROM adidas_us_sales_datasets
WHERE Region = 'Northeast' AND State = 'New York'
ORDER BY Invoice_Date ASC;

--Count products sold by each store
SELECT `Retailer`,
      COUNT (`Retailer ID`) AS Store_sales
FROM adidas_us_sales_datasets
GROUP BY Retailer;

```
